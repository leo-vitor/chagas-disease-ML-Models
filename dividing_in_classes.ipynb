Código para separar os espectrogramas em classes de acordo com os dados da planilha.

```python
import pandas as pd

caminho_planilha = "/home/leo-vitor/Documentos/UFC/biosinais/dataset-pesquisa-ecg/BancoChagas-Geral-Padronizado-25-09-2023.xlsx"

df = pd.read_excel(caminho_planilha)

# Converta a coluna 'FE Teicholz' para strings e, em seguida, para valores numéricos
df["FE Teicholz"] = df["FE Teicholz"].astype(str).str.rstrip("%").astype("float")
print(df["FE Teicholz"])
```

As classes são criadas baseadas nos valores da colunas FE Teicholz, que mede a fração de ejeção do paciente no exame.Dividiremos o conjunto de dados em 5 classes:
- 0-20%
- 21-40%
- 41-60%
- 61-80%
- 81%-100%

```python
# Classe de 0-20%
Classe_1 = (df[df["FE Teicholz"]<=0.2]["filename"].dropna().astype(str).tolist())
print(Classe_1)
# Classe de 21-40%
Classe_2 = df[(df["FE Teicholz"] > 0.2) & (df["FE Teicholz"] <= 0.4)]["filename"].dropna().astype(str).tolist()
print(Classe_2)
# Classe de 41-60%
Classe_3 = df[(df["FE Teicholz"] > 0.4) & (df["FE Teicholz"] <= 0.6)]["filename"].dropna().astype(str).tolist()
print(Classe_3)
# Classe de 61-80%
Classe_4 = df[(df["FE Teicholz"] > 0.6) & (df["FE Teicholz"] <= 0.8)]["filename"].dropna().astype(str).tolist()
print(Classe_4)
# Classe de 81%-100%
Classe_5 = df[df["FE Teicholz"] > 0.8]["filename"].dropna().astype(str).tolist()
print(Classe_5)
```

## Formatando os valores para que eles sejam compativeis com os nomes dos arquivos .txt

Convertendo-os em inteiros, preenchendo-os e os transformando em strings

```python
# Lista de todas as classes
classes = [Classe_1, Classe_2, Classe_3, Classe_4, Classe_5]

# Converter para inteiros, remover '.0', formatar e transformar em listas de strings
listas_formatadas = []

for classe in classes:
    lista_formatada = [str(numero).replace('.0', '').zfill(3) if len(str(numero).replace('.0', '')) < 3 else str(numero).replace('.0', '') for numero in classe]
    listas_formatadas.append(lista_formatada)

print(listas_formatadas)
```

```python
import os
import shutil

# Função para verificar se uma pasta existe e criar se necessário
def criar_pasta(pasta):
    if not os.path.exists(pasta):
        os.makedirs(pasta)

# Função para verificar se um arquivo já foi copiado
def arquivo_copiado(origem, destino):
    return os.path.exists(os.path.join(destino, os.path.basename(origem)))

# Definir pastas de entrada e saída
pasta_entrada = "/home/leo-vitor/Documentos/UFC/biosinais/escalogramas"
pasta_saida = "/home/leo-vitor/Documentos/UFC/biosinais"

# Criar pastas para cada classe, se necessário
pastas_classes = []
for i, classe in enumerate(listas_formatadas, start=1):
    classe_folder = os.path.join(pasta_saida, f"Classe_{i}")
    criar_pasta(classe_folder)
    pastas_classes.append(classe_folder)

# Copiar arquivos para pastas correspondentes
for arquivo in os.listdir(pasta_entrada):
    # Obtém o nome do arquivo sem a extensão
    nome_arquivo, extensao = os.path.splitext(arquivo)

    # Separa o nome do arquivo antes do "-"
    if "-" in nome_arquivo:
        nome_base = nome_arquivo.split("-")[0]
    else:
        nome_base = nome_arquivo

    # Verifica se o nome do arquivo está na lista
    for i, classe in enumerate(listas_formatadas, start=1):
        if nome_base in classe:
            # Constrói o caminho completo do arquivo
            arquivo_completo = os.path.join(pasta_entrada, arquivo)

            # Copia o arquivo para a pasta de saída correspondente à classe, se ainda não foi copiado
            classe_folder = pastas_classes[i - 1]  # Pegar a pasta correspondente à classe
            if not arquivo_copiado(arquivo_completo, classe_folder):
                shutil.copy2(arquivo_completo, classe_folder)
                print(f"Arquivo {arquivo} copiado para {classe_folder}")
            else:
                print(f"Arquivo {arquivo} já foi copiado para {classe_folder}")

            # Liberar memória
            del arquivo_completo
            break  # Se o arquivo for encontrado na classe, não é necessário continuar procurando nas outras classes

# Liberar memória
del pastas_classes
```

### Criando uma relação entre os arquivos e seus prontuários


```python
import pandas as pd
import re

def criar_relacao_filename_prontuario():
    # Caminho do arquivo Excel
    arquivo_excel = "/home/leo-vitor/Documentos/UFC/biosinais/dataset-pesquisa-ecg/BancoChagas-Geral-Padronizado-25-09-2023.xlsx"

    # Ler o arquivo Excel
    planilha = pd.read_excel(arquivo_excel)

    # Criar um dicionário para armazenar a relação filename-prontuário
    relacao_filename_prontuario = {}

    # Para cada linha na planilha
    for index, row in planilha.iterrows():
        # Verificar se o prontuário é válido
        if pd.notnull(row["Prontuario"]):
            # Verificar se filename é uma lista
            if isinstance(row["filename"], list):
                # Para cada filename na lista de filenames
                for filename in row["filename"]:
                    # Verificar se o filename não é NaN
                    if pd.notnull(filename):
                        try:
                            # Formatando o filename
                            formatted_filename = str(filename).replace('.0', '').zfill(3) if len(str(filename).replace('.0', '')) < 3 else str(filename).replace('.0', '')
                            # Adicionar a relação filename-prontuário ao dicionário
                            relacao_filename_prontuario[formatted_filename] = row["Prontuario"]
                        except Exception as e:
                            print(f"Erro ao processar o arquivo {filename}: {e}")
                            continue
            else:
                # Verificar se o filename não é NaN
                if pd.notnull(row["filename"]):
                    try:
                        # Formatando o filename
                        formatted_filename = str(row["filename"]).replace('.0', '').zfill(3) if len(str(row["filename"]).replace('.0', '')) < 3 else str(row["filename"]).replace('.0', '')
                        # Adicionar a relação filename-prontuário ao dicionário
                        relacao_filename_prontuario[formatted_filename] = row["Prontuario"]
                    except Exception as e:
                        print(f"Erro ao processar o arquivo {row['filename']}: {e}")
                        continue

    return relacao_filename_prontuario


# Criar a relação filename-prontuário
relacao_filename_prontuario = criar_relacao_filename_prontuario()
print(relacao_filename_prontuario)
```

Separando os arquivos de escalograma por prontuário, para garantir que escalogramas de um mesmo paciênte não estejam nos conjuntos de teste e treino ao mesmo tempo

```python
import os
import shutil

def organizar_arquivos_por_prontuario(pasta_origem, relacao_filename_prontuario):
    # Pasta de destino para arquivos sem prontuário
    pasta_destino_sem_prontuario = os.path.join(pasta_origem, "Sem_Prontuario")

    # Criar a pasta de destino sem prontuário se ainda não existir
    if not os.path.exists(pasta_destino_sem_prontuario):
        os.makedirs(pasta_destino_sem_prontuario)

    # Iterar sobre os arquivos na pasta de origem
    for arquivo in os.listdir(pasta_origem):
        # Obter o caminho completo do arquivo de origem
        caminho_origem = os.path.join(pasta_origem, arquivo)

        # Verificar se o arquivo existe
        if os.path.isfile(caminho_origem):
            # Verificar se o arquivo já existe na pasta de destino
            if os.path.exists(os.path.join(pasta_destino_sem_prontuario, arquivo)):
                print(f"O arquivo {arquivo} já foi transferido para a pasta de destino.")
                continue

            # Obter o filename sem a extensão
            filename_sem_extensao, _ = os.path.splitext(arquivo)

            # Obter o prefixo do filename (antes do '-')
            prefixo_filename = filename_sem_extensao.split('-')[0]

            # Verificar se o prefixo do filename está na relação filename-prontuário
            if prefixo_filename in relacao_filename_prontuario:
                # Obter o prontuário correspondente ao prefixo do filename
                prontuario = str(relacao_filename_prontuario[prefixo_filename])

                # Criar o caminho da pasta de destino
                pasta_destino = os.path.join(pasta_origem, prontuario)

                # Criar a pasta de destino se ainda não existir
                if not os.path.exists(pasta_destino):
                    os.makedirs(pasta_destino)

                # Mover o arquivo para a pasta de destino
                caminho_destino = os.path.join(pasta_destino, arquivo)
                try:
                    shutil.move(caminho_origem, caminho_destino)
                except FileNotFoundError:
                    print(f"O arquivo {arquivo} não foi encontrado no diretório de origem.")
            else:
                # Mover o arquivo para a pasta de destino sem prontuário
                caminho_destino = os.path.join(pasta_destino_sem_prontuario, arquivo)
                try:
                    shutil.move(caminho_origem, caminho_destino)
                except FileNotFoundError:
                    print(f"O arquivo {arquivo} não foi encontrado no diretório de origem.")
        else:
            print(f"O arquivo {arquivo} não foi encontrado no diretório de origem.")

    print("Arquivos organizados por prontuário com sucesso!")


# Pasta de origem dos arquivos
pasta_classe_1 = "/home/leo-vitor/Documentos/UFC/biosinais/Classe_1"
pasta_classe_2 = "/home/leo-vitor/Documentos/UFC/biosinais/Classe_2"
pasta_classe_3 = "/home/leo-vitor/Documentos/UFC/biosinais/Classe_3"
pasta_classe_4 = "/home/leo-vitor/Documentos/UFC/biosinais/Classe_4"
pasta_classe_5 = "/home/leo-vitor/Documentos/UFC/biosinais/Classe_5"
# Separar os arquivos da pasta em subpastas com o nome do prontuário
organizar_arquivos_por_prontuario(pasta_classe_1, relacao_filename_prontuario)
organizar_arquivos_por_prontuario(pasta_classe_2, relacao_filename_prontuario)
organizar_arquivos_por_prontuario(pasta_classe_3, relacao_filename_prontuario)
organizar_arquivos_por_prontuario(pasta_classe_4, relacao_filename_prontuario)
organizar_arquivos_por_prontuario(pasta_classe_5, relacao_filename_prontuario)
```

## Agora que separamos os dados em suas respectivas classes e de acordo com o número de prontuário, vamos separa-los nos conjuntos de treino e teste.

```python
import os
import shutil
import random

def dividir_dados_por_paciente(caminho_origem, caminho_destino_treino, caminho_destino_teste, porcentagem_treino=0.8):
    # Lista todos os diretórios (pacientes) na pasta de origem
    pacientes = [d for d in os.listdir(caminho_origem) if os.path.isdir(os.path.join(caminho_origem, d))]
    
    # Seleciona aleatoriamente os pacientes para treino e teste
    pacientes_selecionados = random.sample(pacientes, int(len(pacientes) * porcentagem_treino))
    pacientes_teste = [p for p in pacientes if p not in pacientes_selecionados]
    
    # Move os diretórios selecionados para as pastas de treino e teste
    for paciente in pacientes_selecionados:
        shutil.move(os.path.join(caminho_origem, paciente), os.path.join(caminho_destino_treino, paciente))
    for paciente in pacientes_teste:
        shutil.move(os.path.join(caminho_origem, paciente), os.path.join(caminho_destino_teste, paciente))

# Classe 1
caminho_origem_classe_1 = '/home/leo-vitor/Documentos/UFC/biosinais/Classe_1'
caminho_destino_treino_classe_1 = '/home/leo-vitor/Documentos/UFC/biosinais/treino_dir/Classe_1'
caminho_destino_teste_classe_1 = '/home/leo-vitor/Documentos/UFC/biosinais/teste_dir/Classe_1'

dividir_dados_por_paciente(caminho_origem_classe_1, caminho_destino_treino_classe_1, caminho_destino_teste_classe_1)

# Classe 2
caminho_origem_classe_2 = '/home/leo-vitor/Documentos/UFC/biosinais/Classe_2'
caminho_destino_treino_classe_2 = '/home/leo-vitor/Documentos/UFC/biosinais/treino_dir/Classe_2'
caminho_destino_teste_classe_2 = '/home/leo-vitor/Documentos/UFC/biosinais/teste_dir/Classe_2'

dividir_dados_por_paciente(caminho_origem_classe_2, caminho_destino_treino_classe_2, caminho_destino_teste_classe_2)

# Classe 3
caminho_origem_classe_3 = '/home/leo-vitor/Documentos/UFC/biosinais/Classe_3'
caminho_destino_treino_classe_3 = '/home/leo-vitor/Documentos/UFC/biosinais/treino_dir/Classe_3'
caminho_destino_teste_classe_3 = '/home/leo-vitor/Documentos/UFC/biosinais/teste_dir/Classe_3'

dividir_dados_por_paciente(caminho_origem_classe_3, caminho_destino_treino_classe_3, caminho_destino_teste_classe_3)

# Classe 4
caminho_origem_classe_4 = '/home/leo-vitor/Documentos/UFC/biosinais/Classe_4'
caminho_destino_treino_classe_4 = '/home/leo-vitor/Documentos/UFC/biosinais/treino_dir/Classe_4'
caminho_destino_teste_classe_4 = '/home/leo-vitor/Documentos/UFC/biosinais/teste_dir/Classe_4'

dividir_dados_por_paciente(caminho_origem_classe_4, caminho_destino_treino_classe_4, caminho_destino_teste_classe_4)

# Classe 5
caminho_origem_classe_5 = '/home/leo-vitor/Documentos/UFC/biosinais/Classe_5'
caminho_destino_treino_classe_5 = '/home/leo-vitor/Documentos/UFC/biosinais/treino_dir/Classe_5'
caminho_destino_teste_classe_5 = '/home/leo-vitor/Documentos/UFC/biosinais/teste_dir/Classe_5'

dividir_dados_por_paciente(caminho_origem_classe_5, caminho_destino_treino_classe_5, caminho_destino_teste_classe_5)

```
